```python
import pandas as pd
```

```python
enade_en_MCQ = pd.read_csv("enade_en_MCQ_2024_06_27_03_07_11.csv")
enade_pt_MCQ = pd.read_csv("enade_pt_MCQ_2024_06_27_03_07_11.csv")
enem_en_MCQ = pd.read_csv("enem_en_MCQ_2024_06_27_03_07_11.csv")
enem_pt_MCQ = pd.read_csv("enem_pt_MCQ_2024_06_27_03_07_11.csv")
```

```python
enade_en_new = pd.read_csv("enade_en_MCQ_2024_06_27_21_20_56.csv")
enade_pt_new = pd.read_csv("enade_pt_MCQ_2024_06_27_21_20_56.csv")
enem_en_new = pd.read_csv("enem_en_MCQ_2024_06_27_21_20_56.csv")
enem_pt_new = pd.read_csv("enem_pt_MCQ_2024_06_27_21_20_56.csv")
```

```python
enade_en_MCQ_merged = pd.merge(enade_en_MCQ, enade_en_new[['id', 'area', 'year', 'response_gpt_4', 'runtime_gpt_4']], on=['id', 'area', 'year'], suffixes=('', '_y'))
enade_pt_MCQ_merged = pd.merge(enade_pt_MCQ, enade_pt_new[['id', 'area', 'year', 'response_gpt_4', 'runtime_gpt_4']], on=['id', 'area', 'year'], suffixes=('', '_y'))
enem_en_MCQ_merged = pd.merge(enem_en_MCQ, enem_en_new[['id', 'area', 'year', 'response_gpt_4', 'runtime_gpt_4']], on=['id', 'area', 'year'], suffixes=('', '_y'))
enem_pt_MCQ_merged = pd.merge(enem_pt_MCQ, enem_pt_new[['id', 'area', 'year', 'response_gpt_4', 'runtime_gpt_4']], on=['id', 'area', 'year'], suffixes=('', '_y'))
```

```python
enade_en_MCQ_merged.to_csv(f'enade_en_MCQ.csv', index=False)
enade_pt_MCQ_merged[~((enade_pt_MCQ_merged['id']==26) &(enade_pt_MCQ_merged['area']=='Jornalismo'))].to_csv(f'enade_pt_MCQ.csv', index=False)
enem_en_MCQ_merged.to_csv(f'enem_en_MCQ.csv', index=False)
enem_pt_MCQ_merged.to_csv(f'enem_pt_MCQ.csv', index=False)
```
