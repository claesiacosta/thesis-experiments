```python
# Padronizar os valores dos níveis de dificuldade para inglês
df_MCQ_generated['difficult_level'] = df_MCQ_generated['difficult_level'].map({
    'Iniciante': 'Beginner',
    'Intermediário': 'Intermediate',
    'Avançado': 'Advanced'
})

# Padronizar os valores dos níveis de Bloom para inglês
df_MCQ_generated['bloom_level'] = df_MCQ_generated['bloom_level'].map({
    'Analisar': 'Analyze',
    'Análise': 'Analyze',
    'Aplicar': 'Apply',
    'Aplicação': 'Apply',
    'Avaliar': 'Evaluate',
    'Avaliação': 'Evaluate',
    'Compreender': 'Understand',
    'Compreensão': 'Understand',
    'Lembrar': 'Remember',
    'Recordar': 'Remember'
})
```

```python
equal_count = (df_MCQ_generated['correct_answer'] == df_MCQ_generated['system_answer']).sum()
print(equal_count)
not_equal_count = len(df_MCQ_generated) - equal_count
print(not_equal_count)
```

```python
bloom_level_summary = df_MCQ_generated['bloom_level'].value_counts().sort_index()
print("Bloom Level Summary: ", bloom_level_summary)
difficult_level_summary = df_MCQ_generated['difficult_level'].value_counts().sort_index()
print("Difficult Level Summary: ", difficult_level_summary)
relevance_summary = df_MCQ_generated['relevance'].value_counts().sort_index()
print("Relevance Summary: ", relevance_summary)
adherence_summary = df_MCQ_generated['adherence'].value_counts().sort_index()
print("Adherence Summary: ", adherence_summary)
grammar_summary = df_MCQ_generated['grammar'].value_counts().sort_index()
print("Grammar Summary: ", grammar_summary)
answerability_summary = df_MCQ_generated['answerability'].value_counts().sort_index()
print("Answerability Summary: ", answerability_summary)
correctness_summary = df_MCQ_generated['correctness'].value_counts().sort_index()
print("Correctness Summary: ", correctness_summary)
```
